# Handling Missing data

```python
#calculating total missing values in data
total_missing = titanic_df.isnull().sum()
percentge = total_missing/titanic_df.isnull().count()
NAs = pd.concat([total_missing,percentge],axis=1,keys=('Total','Percentge'))
NAs[NAs.Total>0].sort_values(by='Percentge',ascending=True)
```

```python
#we will replace the missing value of Age with its mean
titanic_df['Age'].fillna(titanic_df.groupby('Pclass')['Age'].transform('mean'),inplace = True)
```

```python
titanic_df.isnull().sum()
```

```python
df=titanic_df
```

```python
#Cabin col has many null values and we do not need cabin col for our model so we will remove this col
df.drop("Cabin",inplace=True,axis=1)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
#numeric data types
df_num = df.select_dtypes(include=[np.number]).columns.tolist()
df_num = df[['PassengerId', 'Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
df_num
```

# Creating dummies

```python
df_cat = df[['Name', 'Sex', 'Ticket', 'Embarked']]
```

```python
#creating dummies of categorical data
df_cat_dummies = pd.get_dummies(df_cat)
```

```python
#combining the data frame of numerical data and categorical data
combined_df = pd.concat([df_num,df_cat_dummies], axis=1)
```

```python
combined_df.head()
```

```python
combined_df.tail()
```

```python
combined_df.isnull().sum()
```

# Creating model

```python
X = combined_df.drop("Survived",axis=1)
y = combined_df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
# Making predictions using the model
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,roc_auc_score
accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
```

```python
#displaying the results of our logistic model
print('Accuracy is  :' ,accuracy)
print('F1 score is :' ,f1)
print('Precision is  :',precision)
print('Recall is  :',recall)
print('Roc Auc is  :',roc_auc)
```

# RANDOM FOREST 

```python
#import random forest 
from sklearn.ensemble import RandomForestClassifier

# build a classifier
rf = RandomForestClassifier(n_estimators=100)

#Train the model using the training sets
rf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = rf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# Model Precision
print("Precision:",metrics.precision_score(y_test, y_pred))
# Model Recall
print("Recall:",metrics.recall_score(y_test, y_pred))
```
